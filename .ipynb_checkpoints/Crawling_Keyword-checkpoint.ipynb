```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# Selenium WebDriver 초기화
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Chrome 브라우저 열기
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# 웹 페이지 열기
driver.get("https://www.freemoa.net/m7/freelance_detail?fno=31887")

try:

        # 클래스 이름을 사용하여 <pre> 태그 선택
    introduction = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "pre.patnersIntroduction"))
    ).text.strip()

    
    
    # "등록 포트폴리오" 정보 추출
    portfolio_count = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//div[@class='colTextA'][b='등록 포트폴리오']/p/span"))
    ).text.strip()

    # "계약 프로젝트" 정보 추출
    project_count = driver.find_element(By.XPATH, "//div[@class='colTextA'][b='계약 프로젝트']/p/span").text.strip()

    # "누적 계약금" 정보 추출
    total_amount = driver.find_element(By.XPATH, "//div[@class='colTextA'][b='누적 계약금']/p/span").text.strip()

    # "평점" 정보 추출
    rating = driver.find_element(By.CSS_SELECTOR, "div.starRating").get_attribute("data-star")

    # "평가 개수" 정보 추출
    review_count = driver.find_element(By.XPATH, "//div[@class='colTextA']/p[b]/span").text.strip()

    # 출력
    print(f"소개:{introduction}")
    print(f"등록 포트폴리오: {portfolio_count} 개")
    print(f"계약 프로젝트: {project_count} 건")
    print(f"누적 계약금: {total_amount}")
    print(f"평점: {rating}")
    print(f"평가 개수: {review_count} 개")

finally:
    # 브라우저 닫기
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Selenium WebDriver 초기화
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Chrome 브라우저 열기
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# 웹 페이지 열기
driver.get('https://www.freemoa.net/m7/s71')
time.sleep(8) 
# 팝업창 닫기
try:
    close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "div.closes"))
    )
    close_button.click()
except Exception as e:
    print("팝업창을 닫는 데 실패했습니다:", e)

time.sleep(8)  # 페이지가 완전히 로드될 때까지 대기

# 모든 파트너 정보 크롤링
partner_data = []

for i in range(1, 11):  # 첫 10명의 파트너에 대해 정보 수집
    try:
        partner_id_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/div/section/div/p/b'
        favorite_cnt_xpath = f'//*[@id="favoriteCntSpan2438"]'
        introduction_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/section/p'
        rating_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/div[2]/p/b'
        review_count_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/div[2]/p/span'
        contract_count_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/p[1]/b'
        project_count_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/p[2]/b'
        total_amount_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/p[3]/b'

        partner_id = driver.find_element(By.XPATH, partner_id_xpath).text.strip()
        favorite_cnt = driver.find_element(By.XPATH, favorite_cnt_xpath).text.strip()
        introduction = driver.find_element(By.XPATH, introduction_xpath).text.strip()
        rating = driver.find_element(By.XPATH, rating_xpath).text.strip()
        review_count = driver.find_element(By.XPATH, review_count_xpath).text.strip()
        contract_count = driver.find_element(By.XPATH, contract_count_xpath).text.strip()
        project_count = driver.find_element(By.XPATH, project_count_xpath).text.strip()
        total_amount = driver.find_element(By.XPATH, total_amount_xpath).text.strip()

        partner_data.append({
            '파트너 아이디': partner_id,
            '관심개수': favorite_cnt,
            '소개글': introduction,
            '평점': rating,
            '평가개수': review_count,
            '계약 개수': contract_count,
            '프로젝트 개수': project_count,
            '누적금액': total_amount
        })

        # 디버깅 출력
        print(f"파트너 아이디: {partner_id}")
        print(f"관심개수: {favorite_cnt}")
        print(f"소개글: {introduction}")
        print(f"평점: {rating}")
        print(f"평가개수: {review_count}")
        print(f"계약 개수: {contract_count}")
        print(f"프로젝트 개수: {project_count}")
        print(f"누적금액: {total_amount}")
        print("-" * 50)

    except Exception as e:
        print(f"{i}번째 파트너 정보 수집 중 오류 발생: {e}")

# 브라우저 닫기
driver.quit()

# 크롤링된 데이터 출력 또는 저장
print("크롤링된 데이터:")
for data in partner_data:
    print(data)
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Selenium WebDriver 초기화
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Chrome 브라우저 열기
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# 웹 페이지 열기
driver.get('https://www.freemoa.net/m7/s71')
time.sleep(8) 
# 팝업창 닫기
try:
    close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "div.closes"))
    )
    close_button.click()
except Exception as e:
    print("팝업창을 닫는 데 실패했습니다:", e)

time.sleep(8)  # 페이지가 완전히 로드될 때까지 대기


# 모든 파트너 정보 크롤링
partner_data = []
num_partners = 0

while num_partners < 100:
    for i in range(1, 11):  # 페이지에 표시된 10명의 파트너에 대해 정보 수집
        if num_partners >= 100:
            break
        try:
            partner_id_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/div/section/div/p/b'
            favorite_cnt_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/div/section/div/p/span'
            introduction_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/section/p'
            rating_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/div[2]/p/b'
            review_count_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/div[2]/p/span'
            contract_count_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/p[1]/b'
            project_count_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/p[2]/b'
            total_amount_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/p[3]/b'

            partner_id = driver.find_element(By.XPATH, partner_id_xpath).text.strip()
            favorite_cnt = driver.find_element(By.XPATH, favorite_cnt_xpath).text.strip()
            introduction = driver.find_element(By.XPATH, introduction_xpath).text.strip()
            rating = driver.find_element(By.XPATH, rating_xpath).text.strip()
            review_count = driver.find_element(By.XPATH, review_count_xpath).text.strip()
            contract_count = driver.find_element(By.XPATH, contract_count_xpath).text.strip()
            project_count = driver.find_element(By.XPATH, project_count_xpath).text.strip()
            total_amount = driver.find_element(By.XPATH, total_amount_xpath).text.strip()

            partner_data.append({
                '파트너 아이디': partner_id,
                '관심개수': favorite_cnt,
                '소개글': introduction,
                '평점': rating,
                '평가개수': review_count,
                '계약 개수': contract_count,
                '프로젝트 개수': project_count,
                '누적금액': total_amount
            })

            num_partners += 1

            # 디버깅 출력
            print(f"{num_partners}번째 파트너 아이디: {partner_id}")

        except Exception as e:
            print(f"{num_partners + 1}번째 파트너 정보 수집 중 오류 발생: {e}")

    # 더보기 버튼 클릭
    if num_partners < 100:
        try:
            more_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.ID, "moreBtn"))
            )
            more_button.click()
            time.sleep(3)  # 새로운 정보가 로드될 때까지 대기
        except Exception as e:
            print("더보기 버튼을 클릭하는 중 오류 발생:", e)
            break

# 브라우저 닫기
driver.quit()

# 데이터를 pandas DataFrame에 저장
df = pd.DataFrame(partner_data)

# DataFrame 출력
print(df)

# DataFrame을 CSV 파일로 저장 (원하는 경우)
df.to_csv('partner_data.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Selenium WebDriver 초기화
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Chrome 브라우저 열기
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# 웹 페이지 열기
driver.get('https://www.freemoa.net/m7/s71')

# 팝업창 닫기
try:
    close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "div.closes"))
    )
    close_button.click()
except Exception as e:
    print("팝업창을 닫는 데 실패했습니다:", e)

time.sleep(2)  # 페이지가 완전히 로드될 때까지 대기

# 모든 파트너 정보 크롤링
partner_data = []
page_number = 1

while True:
    for i in range(1, 11):  # 페이지에 표시된 10명의 파트너에 대해 정보 수집
        try:
            partner_id_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/div/section/div/p/b'
            personal_corp_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/div/section/div/p/span'
            favorite_cnt_xpath = f'//*[@id="favoriteCntSpan31887"]'  # 관심개수 XPath
            introduction_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/section/p'
            rating_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/div[2]/p/b'
            review_count_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/div[2]/p/span'
            contract_count_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/p[1]/b'
            project_count_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/p[2]/b'
            total_amount_xpath = f'//*[@id="freelancerList"]/div[{i}]/div/article/div/p[3]/b'

            partner_id = driver.find_element(By.XPATH, partner_id_xpath).text.strip()
            personal_corp = driver.find_element(By.XPATH, personal_corp_xpath).text.strip()
            favorite_cnt = driver.find_element(By.XPATH, favorite_cnt_xpath).text.strip()
            introduction = driver.find_element(By.XPATH, introduction_xpath).text.strip()
            rating = driver.find_element(By.XPATH, rating_xpath).text.strip()
            review_count = driver.find_element(By.XPATH, review_count_xpath).text.strip()
            contract_count = driver.find_element(By.XPATH, contract_count_xpath).text.strip()
            project_count = driver.find_element(By.XPATH, project_count_xpath).text.strip()
            total_amount = driver.find_element(By.XPATH, total_amount_xpath).text.strip()

            partner_data.append({
                '파트너 아이디': partner_id,
                '개인/법인': personal_corp,
                '관심개수': favorite_cnt,
                '소개글': introduction,
                '평점': rating,
                '평가개수': review_count,
                '계약 개수': contract_count,
                '프로젝트 개수': project_count,
                '누적금액': total_amount
            })

            # 디버깅 출력
            print(f"{len(partner_data)}번째 파트너 아이디: {partner_id}")

        except Exception as e:
            print(f"파트너 정보 수집 중 오류 발생: {e}")

    # 더보기 버튼 클릭 시도
    try:
        more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.ID, "moreBtn"))
        )
        more_button.click()
        page_number += 1
        print(f"{page_number}번째 페이지 로드 중...")
        time.sleep(3)  # 새로운 정보가 로드될 때까지 대기
    except Exception as e:
        print("더보기 버튼이 더 이상 존재하지 않거나 오류가 발생했습니다:", e)
        break

# 브라우저 닫기
driver.quit()

# 데이터를 pandas DataFrame에 저장
df = pd.DataFrame(partner_data)

# DataFrame 출력
print(df)

# DataFrame을 CSV 파일로 저장 (원하는 경우)
df.to_csv('partner_data.csv', index=False, encoding='utf-8-sig')
```

### 전처리

```python
df['사업자']=df['개인/법인'].apply(lambda x:x.split('|')[0] if '|' in x else x)
```

```python
df.head()
```

```python
df['지역']=df['개인/법인'].apply(lambda x: x.split('|')[1] if '|' in x else 'none')
```

```python
df['소개글길이']=df['소개글'].apply(lambda x: len(x) )
```

```python
df.head()
```

```python
df.columns
```

```python
df['계약 개수']=df['계약 개수'].str[:-1]
```

```python
df['프로젝트 개수']=df['프로젝트 개수'].str[:-1]
```

```python
df['누적금액']=df['누적금액'].str[:-2]
```

```python
df.info()
```

```python
df['프로젝트 개수']=df['프로젝트 개수'].astype('int64')
df['누적금액'] = df['누적금액'].str.replace(',', '').astype('int64')
df['계약 개수']=df['계약 개수'].astype('int64')
```

```python
df.drop('관심개수',axis=1, inplace=True)
```
